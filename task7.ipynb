```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

def func(spl):
    tf = TfidfVectorizer(smooth_idf=False, sublinear_tf=False, norm=None, analyzer='word')
    txt_fitted = tf.fit(spl)
    txt_transformed = txt_fitted.transform(spl)
    tf.vocabulary_
    idf = tf.idf_
    dict(zip(txt_fitted.get_feature_names(), idf))
    # get feature names
    feature_names = np.array(tf.get_feature_names())
    sorted_by_idf = np.argsort(tf.idf_)
    new1 = tf.transform(spl)
    # find maximum value for each of the features over all of dataset:
    max_val = new1.max(axis=0).toarray().ravel()
    #sort weights from smallest to biggest and extract their indices 
    sort_by_tfidf = max_val.argsort()
    print("\nFeatures with highest tfidf: \n{}".format(
      feature_names[sort_by_tfidf[-10:]]))
```

```python
def bigrams(seq, n):
    return list(zip(*(seq[index:] for index in range(n))))

bigrams(words[:10], 2)
```

```python
from operator import *

def flatten_corpus(corpus):
    return ' '.join([document.strip() for document in corpus])

def topbigr(corpus, bigr_val=2, limit=100):
    corpus = flatten_corpus(corpus)
    tokens = nltk.word_tokenize(corpus)
    
    bigram = bigrams(tokens, bigr_val)
    bigr_freq_dist = nltk.FreqDist(bigram)
    sorted_bigram_fd = sorted(bigr_freq_dist.items(), key=itemgetter(1), reverse=True)
    
    sorted_bigram = sorted_bigram_fd[0:limit]
    sorted_bigram = [(' '.join(text), freq) for text, freq in sorted_bigram]
    return sorted_bigram

bigrAlice = topbigr(corpus=words)
```
